Load up the sympy python package to do symbolic mathematics 

```python
from IPython.display import display

from sympy.interactive import printing
printing.init_printing()

from __future__ import division
import sympy as sym



import matplotlib
matplotlib.rc('font',**{'family':'serif'})
```

```python
E0    = sym.symbols(r"\mathcal{E}_{0}", cls=sym.Function, real=True)
k     = sym.symbols(r"k", real=True)
l     = sym.symbols(r"\ell", real=True)
om    = sym.symbols(r"\omega", real=True)
t     = sym.symbols(r"t", real=True)
vphi  = sym.symbols(r"\varphi", cls=sym.Function, real=True)
alpha = sym.symbols(r"\alpha", cls=sym.Function, real=True)
phi   = sym.symbols(r"\phi", cls=sym.Function, real=True)

Efield1 = E0(l)*sym.exp(sym.I*(k*l - om*t + vphi(l)))* sym.cos(alpha(l)) 
Efield2 = E0(l)*sym.exp(sym.I*(k*l - om*t + vphi(l)))* sym.sin(alpha(l)) *sym.exp(sym.I*phi(l))

display(Efield1)
display(Efield2)
                                         
```

```python
d11 = sym.simplify(sym.diff( Efield1, l, 1) )
d12 = sym.simplify(sym.diff( Efield1, l, 2) )
display(Efield1)
display(d11)
display(d12)
display(sym.simplify(d12 + k**2 * Efield1))
```

```python
d21 = sym.simplify( sym.diff( Efield2, l, 1) )
d22 = sym.simplify( sym.diff( Efield2, l, 2) )
display(Efield2)
display(d21)
display(d22)
display(sym.simplify(d22 + k**2 * Efield2))
```

```python
k     = sym.symbols(r"k", real=True)
l     = sym.symbols(r"\ell", real=True)
om    = sym.symbols(r"\omega", real=True)
t     = sym.symbols(r"t", real=True)
E0    = sym.symbols(r"\mathcal{E}_{0}", real=True)
vphi  = sym.symbols(r"\varphi", real=True)
alpha = sym.symbols(r"\alpha", real=True)
phi   = sym.symbols(r"\phi", real=True)

dE0    = sym.symbols(r"\mathcal{E'}_{0}", real=True)
dvphi  = sym.symbols(r"\varphi'", real=True)
dalpha = sym.symbols(r"\alpha'", real=True)
dphi   = sym.symbols(r"\phi'", real=True)

P1 = sym.symbols(r"P_{1}", real=True)
P2 = sym.symbols(r"P_{2}", real=True)
P3 = sym.symbols(r"P_{3}", real=True)
P4 = sym.symbols(r"P_{4}", real=True)

re1=-2*k*E0*sym.cos(alpha)*dvphi + 4*sym.pi*k**2 * P1
im1=-2*k*E0*sym.sin(alpha)*dalpha + 2*k*sym.cos(alpha)*dE0 - 4*sym.pi*k**2 * P2

re2=-2*k*E0*sym.sin(alpha)*dphi - 2*k*E0*sym.sin(alpha)*dvphi + 4*sym.pi*k**2 * P3
im2=2*k*E0*sym.cos(alpha)*dalpha + 2*k*sym.sin(alpha)*dE0 - 4*sym.pi*k**2 * P4 

display(re1)
display(im1)
display(re2)
display(im2)

equations = [ re1, im1, re2, im2 ] 
variables = [dE0, dvphi, dalpha, dphi]
sol = sym.solve(equations,variables, set=True)


```
