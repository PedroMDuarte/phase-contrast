```python
from sympy.physics.quantum.cg import Wigner3j

def dipole( q, ns, Js, mjs  ):
    #print
    #print "q=",q
    #print ns
    def f(i,j, p=False):
        if ns[0,i,j] == ns[1,i,j]:
            return 0
        elif np.abs(ns[0,i,j] - ns[1,i,j]) == 1.:
            # This is the sign on the Wigner-Eckart theorem
            sign = int(np.power(-1, Js[0,i,j]-mjs[0,i,j] ))
            
            # This is the sign due to the order of the bra-ket on the
            # reduced matrix element
            if ns[1,i,j] > ns[0,i,j]:
                #ordering is correct
                red_sign = 1 
            else:
                #ordering is flipped
                red_sign = int(np.power( -1, Js[1,i,j]-Js[0,i,j] ))
        
            wigner3 = Wigner3j( Js[0,i,j], -mjs[0,i,j], 1, q,  Js[1,i,j], mjs[1,i,j] ).doit()
            if p:
                print "sign=",sign
                print "red_sign=", red_sign
                print '{0:.2f}, {1:.2f}, {2:.2f}, {3:.2f}, {4:.2f}, {5:.2f}'.format( Js[0,i,j], -mjs[0,i,j], 1, q,  Js[1,i,j], mjs[1,i,j] )
                print "wigner=",wigner3
            return sign*wigner3*red_sign
        else:
            return 0
    mdim = len(ns[0])
    norm = 1e3
    for i in range(mdim):
        for j in range(mdim):
            #print i,j
            #fij = f(i,j,p=True)
            #print "fij =",fij
            fij = np.abs(f(i,j))
            if fij>0 and fij < norm:
                norm = fij
    matrix = sym.Matrix(mdim, mdim, f)/norm
    return q, norm, matrix

def get_qnumbers( Jg, Je, mJg, mJe):
    if 2*Jg % 1. != 0:
        print Jg
        print 2*Jg
        print 2*Jg % 1
        raise ValueError("Invalid value for total angular momentum Jg")
    if 2*Je % 1. != 0:
        print Je
        print 2*Je
        print 2*Je % 1
        raise ValueError("Invalid value for total angular momentum Je")
    
    if mJg is None:
        mJg = np.arange(-Jg,Jg+0.1,1)
    else:
        mJg = np.array(mJg)
    if mJe is None:
        mJe = np.arange(-Je,Je+0.1,1)
    else:
        mJe = np.array(mJe)
        
    #print "Inside get_qnumbers:"
    #print mJg
    #print mJe
    mJs =  np.array(np.meshgrid( np.hstack((mJg,mJe)), np.hstack((mJg,mJe)) ))
    
    nge = np.hstack((np.zeros( len(mJg)),np.ones(len(mJe)) ))
    ns = np.array(np.meshgrid( nge, nge), dtype=int)
    
    Jge = np.hstack(( np.ones(len(mJg))*Jg, np.ones(len(mJe))*Je ))
    Js = np.array(np.meshgrid( Jge, Jge ))

    # The arrays are reversed so that in the meshgrid
    # index 0 = row index
    # index 1 = col index
    return ns[::-1], Js[::-1], mJs[::-1]

def dipole_operator(Jg, Je, mJg=None, mJe=None):
    ns, Js, mjs = get_qnumbers( Jg, Je, mJg, mJe)
    
    #display(mjs)
    #display(ns)
    #display(Js)
    dqs = []
    for q in [-1, 0, +1]:
        dqs.append( dipole( q, ns, Js, mjs  ) ) 
    return dqs

def format_mJ(n, mJ):
    if n == 0:
        n = 'g'
    elif n==1:
        n = 'e'
    else:
        raise ValueError("Undefined quantum level is not ground or excited.")
        
    if mJ % 1 == 0 :
        mJ = r'%s(%d)'%(n,mJ)
    elif mJ % 0.5 == 0:
        #mJ = r'g_{\frac{%d}{%d}}' % (mJ/0.5,2)
        mJ = r'%s(\frac{%d}{%d})' % (n,mJ/0.5,2)
    else:
        raise ValueError("mJ is not integer or half-integer.")
    return mJ

def atomic_density_matrix(Jg, Je, mJg=None, mJe=None):
    ns, Js, mjs = get_qnumbers( Jg, Je, mJg, mJe)
    
    mdim = len(mjs[0])
    rho_symbolsRe = []
    rho_symbolsIm = []
    for i in range(mdim):
        rowRe = sym.symbols(
                  ' '.join( [r"Re\rho_{%s%s}"% \
                            (format_mJ(ns[0,i,k], mjs[0,i,k]), format_mJ(ns[1,i,k],mjs[1,i,k]) )\
                            for k in range(mdim) ])\
                          ,real=True)
        rowIm = sym.symbols(
                  ' '.join( [r"Im\rho_{%s%s}"% \
                            (format_mJ(ns[0,i,k], mjs[0,i,k]), format_mJ(ns[1,i,k],mjs[1,i,k]) )\
                            for k in range(mdim) ])\
                          ,real=True)
        rho_symbolsRe.append( rowRe ) 
        rho_symbolsIm.append( rowIm ) 
    def f(i,j):
        if i<j:
            return rho_symbolsRe[i][j] + I*rho_symbolsIm[i][j]
        elif i==j:
            return rho_symbolsRe[i][j]
        else:
            return sym.conjugate( rho_symbolsRe[j][i] + I*rho_symbolsIm[j][i] ) 
    matrix = sym.Matrix(mdim,mdim, f)
    return matrix

def H0(omega, Jg, Je, mJg=None, mJe=None):
    ns, Js, mjs = get_qnumbers( Jg, Je, mJg, mJe)
    mdim = len(mjs[0])
    #print "Inside H0, size of matrix =",mdim
    diag = []
    #diag = ''
    for k in range(mdim):
        if ns[0,k,k] == 0:
            diag.append( 0 ) 
            #diag = diag + '0 ' 
        else:
            detuning = sym.symbols(r"\Delta_{%s} "%format_mJ(ns[0,k,k], mjs[0,k,k]), real=True)
            diag.append( omega - detuning )
            #diag.append( sym.symbols( r"\omega_{%s} "%format_mJ(ns[0,k,k], mjs[0,k,k]), real=True) )
            
            
    #diag = sym.symbols(diag)
        
    def f(i,j):
        if i != j:
            return 0
        else:
            return diag[i] 

    matrix = sym.Matrix(mdim, mdim, f)
    return matrix

# ROTATION TO THE ROTATING FRAME

def Utransf(omega, t, Jg, Je, mJg=None, mJe=None):
    ns, Js, mjs = get_qnumbers( Jg, Je, mJg, mJe)
    #print ns
    mdim = len(mjs[0])
    #print "Inside Utransf, size of matrix =",mdim
    diag = []
    for k in range(mdim):
        if ns[0,k,k] == 0:
            diag.append(0)
        else:
            diag.append(omega)
    #print diag
    def f(i,j):
        if i != j:
            return 0
        else:
            if diag[i] == 0:
                return 1
            else:
                return sym.exp(-I*diag[i]*t)
    Umatrix = sym.Matrix(mdim, mdim, f)
    
    def fcc(i,j):
        return sym.conjugate( f(j,i) ) 
    Uccmatrix = sym.Matrix(mdim,mdim,fcc)
    
    def fdu(i,j):
        return sym.diff( f(i,j), t)
    dU = sym.Matrix(mdim,mdim, fdu)
    
    return Umatrix, Uccmatrix, dU

# INTERACTION

def HI(Ox,Oy,Oz,omega,t,Jg, Je, mJg=None, mJe=None ):
    ns, Js, mjs = get_qnumbers( Jg, Je, mJg, mJe)
    dqs={}
    for q in [-1, 0, +1]:
        d = dipole( q, ns, Js, mjs  ) 
        dqs[q] = d[1]*d[2]
    #for q in dqs.keys():
    #    print "q=",q
    #    display(dqs[q])
    dx = 1/sym.sqrt(2) * ( dqs[-1] - dqs[+1])
    dy = I/sym.sqrt(2) * ( dqs[-1] + dqs[+1])
    dz = dqs[0] 
    #print "dx, dy, dz = "
    #display( dx, dy, dz)
    Hint =  sym.sqrt(3)*(-Ox*dx - Oy*dy - Oz*dz)*(sym.exp(I*omega * t ) + sym.exp(-I*omega * t ) )/2
    Hint.simplify()
    return Hint

def rotating_wave_approx( Ht ):
    Ht = Ht.applyfunc(lambda x: x.expand().collect(sym.exp(-2*I*omega*t)).coeff(sym.exp(-2*I*omega*t),0) )
    Ht = Ht.applyfunc(lambda x: x.expand().collect(sym.exp(+2*I*omega*t)).coeff(sym.exp(+2*I*omega*t),0) )
    return Ht


# RELAXATION 

def relax(gamma, Jg, Je, mJg=None, mJe=None):
    ns, Js, mjs = get_qnumbers( Jg, Je, mJg, mJe)
    #print ns
    mdim = len(mjs[0])
    def f(i,j):
        if i != j:
            return 0
        else:
            if ns[0,i,i] == 0:
                return 0
            else:
                return gamma
    return sym.Matrix(mdim, mdim, f)

# REPOPULATION
# The treatment for repopulation is in section 12.1 in the book
# "Optically polarized atoms", where the spontaneous emission 
# operator is defined

def repop(rho, gamma, Jg, Je, mJg=None, mJe=None):
    ns, Js, mjs = get_qnumbers( Jg, Je, mJg, mJe)
    #print ns
    mdim = len(mjs[0])
    
    gnddim = 0
    for i in range(mdim):
        if ns[0,i,i] == 0:
            gnddim+= 1 
    excdim = mdim - gnddim
    
    def f(m,n):
        if ns[0,m,n] == 0 and ns[1,m,n] == 0:
            rs_sum = 0
            for r in range(mdim):
                for s in range(mdim):
                    if ns[0,r,s] == 1 and ns[1,r,s] == 1:
                        sign = int(np.power(-1, 2*Jg-mjs[0,m,n]-mjs[1,m,n] ))
                        sumq = 0
                        for q in [-1, 0, 1]:
                            wigner3jA = Wigner3j( Jg, -mjs[0,m,n], 1, q,  Je, mjs[1,r,s] ).doit()
                            wigner3jB = Wigner3j( Jg, -mjs[1,m,n], 1, q,  Je, mjs[0,r,s] ).doit()
                            sumq += wigner3jA*wigner3jB
                        Fmnrs =  sign*(int(2*Je) + 1)*gamma * sumq
                        rs_sum = rs_sum + rho[r,s] * Fmnrs
                    else:
                        rs_sum += 0
            return rs_sum
        else:
            return 0
    return sym.Matrix(mdim, mdim, f)
    
```

```python
def eqs_light_atoms( Jg, Je, **kwargs):
    mJg=kwargs.get('mJg',None)
    mJe=kwargs.get('mJe',None)
    
    # Create density matrix for display
    rho = atomic_density_matrix( Jg, Je, mJg=mJg, mJe=mJe)
    print "Atomic density matrix = \n" 
    display(rho)
    print "\n"

    # Create atom part of the Hamiltonian
    # omega is the frequency of the electric field
    omega = sym.symbols(r"\omega", real=True)
    E0 = H0(omega, Jg, Je, mJg=mJg, mJe=mJe)
    print "Atom part of the Hamiltonian = \n" 
    display(E0)
    print "\n" 

    # Create the dipole operators 
    dqs = dipole_operator( Jg, Je, mJg=mJg, mJe=mJe ) 
    print "Dipole operators in spherical basis = \n"
    display(dqs)
    print "\n" 

    # Create the unitary matrix that will be used to 
    # go to the rotating frame
    t = sym.symbols("t",real=True)
    U, Ucc, dU= Utransf(omega, t, Jg, Je, mJg=mJg, mJe=mJe)
    display(U, Ucc, dU)

    # Create the atom + interacion part of the Hamiltonian
    # in the rotating frame and with the rotating wave
    # approximation
    
    # start with the lab frame interaction part
    Ox = kwargs.get('Ox', sym.symbols("\Omega_{x}", real=True))
    Oy = kwargs.get('Oy', sym.symbols("\Omega_{y}", real=True))
    Oz = kwargs.get('Oz', sym.symbols("\Omega_{z}", real=True))
    EI = HI(Ox, Oy, Oz, omega, t, Jg, Je, mJg=mJg, mJe=mJe)
    
    
    
    print "Total Hamiltonian in the lab frame = \n"
    display( E0+EI ) 
    print "\n"
    Ht = Ucc*(E0+EI)*U - I*Ucc*dU
    Ht.expand()
    Ht = rotating_wave_approx(Ht)
    print "Total Hamiltonian in the rotating frame, with RWA= \n" 
    display( Ht )
    print "\n" 
    
    
    # Create the relaxation matrices
    gamma = sym.symbols(r"\Gamma", real=True)
    G = relax(gamma, Jg, Je, mJg=mJg, mJe=mJe)
    print "Relaxation matrix = \n"
    display(G)
    print "\n"
    
    L = repop(rho, gamma, Jg, Je, mJg=mJg, mJe=mJe)
    print "Repopulation rate matrix = \n"
    display(L)
    print "\n"
    
    # Create the set of equations that make up the Liouville equation
    commutator = (Ht*rho - rho*Ht)/I
    decay = -1*(G*rho + rho*G)/2

    #display(commutator)
    #display(decay)
    mdim = commutator.shape[0]
    equations = []
    variables = []
    nconj = 0 
    for i in range(mdim):
        for j in range(mdim):
            # Check if an equation is the complex conjugate of an existing equation
            e = sym.simplify( commutator[i,j] + decay[i,j] + L[i,j] )
            v = rho[i,j]
            add = True
            for k in range(len(equations)):
                econj = equations[k] == sym.conjugate( e ).simplify()
                vconj = variables[k] == sym.conjugate( v ).simplify()
                if econj and vconj:
                    nconj += 1
                    add = False
                elif econj or vconj:
                    print "Adding:"
                    display( [v, e] ) 
                    print "\n"
                    print "currently we have:"
                    for  m in range(len(equations)):
                        display([variables[m],equations[m]])
                    print "equations are conjugate ?", econj
                    print "variables are conjugate ?", vconj
                    raise ValueError("Equation for conjugate variable is not conjugate of equation!!!!\nSomething must be wrong...")
            if add:
                equations.append(e)
                variables.append(v) 


    print "Full Liouville set of equations ="
    print "(%d discarded for being conjugate)\n" % nconj
    for i in range(len(equations)):
        display([variables[i], equations[i]]) 
    print "\n"
    
    print "Full Liouville set, real and imaginary parts = \n"
    eqs = []
    for i in range(len(equations)):
        real = sym.re(equations[i]).simplify()
        imag = sym.im(equations[i]).simplify()
        if real != 0:
            eqs.append( real )
        if imag != 0:
            eqs.append( imag ) 
    var = []
    pops = []
    for i in range(mdim):
        for j in range(mdim):
            if i<j:
                var.append( sym.re(rho[i,j]) ) 
                var.append( sym.im(rho[i,j]) ) 
            elif i == j :
                var.append( sym.re(rho[i,j]) ) 
                pops.append( sym.re(rho[i,j]) ) 
    
    # Construct equation for conservation of the population
    popeq = -1 
    for p in pops:
        popeq = popeq + p
        
    # Retain only linearly independent equations
    indeqs = [popeq]
    for i,eq in enumerate(eqs):
        check = []
        for e in indeqs:
            r = sym.simplify(eq/e)
            check.append( r.is_Number ) 
        #print "checking eq #",i
        #print check
        if True not in check:
            indeqs.append(eq) 
        #else:
        #    print "skipped duplicate", i
    
    print "equations:\n"
    for eq in eqs:
        display(eq)
    print "\n"
    print "linearly independent equations plus population conservation:"
    print "total = %d\n" % len(indeqs) 
    
    for eq in indeqs:
        display(eq)
    print "\n"
    print "variables:"
    print "total = %d\n" % len(var) 
    for v in var:
        display(v) 
    print "\n" 
    
    return indeqs, var
    #sol = sym.solve(equations,variables, set=True)

    
```
